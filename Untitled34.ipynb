```python
# Import necessary libraries
import pandas as pd

# Load the dataset
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# 3.1 Identify inconsistent or irrelevant entries
def check_inconsistent_entries(column_name, top_n=20):
    print(f"\n=== Analyzing '{column_name}' column ===")

    # Check basic stats
    value_counts = df[column_name].value_counts(dropna=False)
    print(f"\nTotal unique values: {len(value_counts)}")
    print(f"Missing values: {df[column_name].isna().sum()} ({df[column_name].isna().mean()*100:.1f}%)")

    # Display most common and least common values
    print("\nMost common values:")
    print(value_counts.head(top_n))

    print("\nLeast common values:")
    print(value_counts.tail(top_n))

    # Check for potential inconsistencies
    if df[column_name].dtype == 'object':
        # Look for case inconsistencies
        case_variations = df[column_name].str.lower().value_counts().head()
        if len(case_variations) < len(value_counts):
            print("\nPotential case inconsistencies (top lowercase variants):")
            print(case_variations)

        # Look for whitespace issues
        whitespace_issues = df[column_name][df[column_name].str.strip() != df[column_name]]
        if not whitespace_issues.empty:
            print(f"\nWhitespace issues found: {len(whitespace_issues)} entries")

        # Look for placeholder/irrelevant values
        common_placeholders = ['n/a', 'na', 'none', 'null', '-', 'unknown']
        placeholders = df[column_name].str.lower().isin(common_placeholders)
        if placeholders.any():
            print(f"\nPotential placeholder values found: {placeholders.sum()} entries")

    return value_counts

# Analyze Country column
country_counts = check_inconsistent_entries('Country')

# Additional checks for other columns of interest
# check_inconsistent_entries('Age')
# check_inconsistent_entries('Employment')

# Optional: Visualize country distribution
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
country_counts.head(20).plot(kind='barh')
plt.title('Top 20 Countries by Response Count')
plt.xlabel('Count')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
# 3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.
```

```python
import pandas as pd

# Load the dataset
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# 3.2 Standardize entries in specific columns
def standardize_column(df, column_name, mapping_dict, case_sensitive=False):
    """
    Standardize values in a column based on a mapping dictionary.

    Parameters:
    - df: DataFrame
    - column_name: Name of column to standardize
    - mapping_dict: Dictionary of {old_value: new_value} mappings
    - case_sensitive: Whether to consider case in matching
    """
    print(f"\nStandardizing '{column_name}' column...")

    # Convert all values to strings for consistent handling
    df[column_name] = df[column_name].astype(str).str.strip()

    # Pre-standardization report
    original_values = df[column_name].value_counts(dropna=False)
    print(f"\nOriginal unique values: {len(original_values)}")

    if not case_sensitive:
        # Convert to lowercase for case-insensitive matching
        df[column_name] = df[column_name].str.lower()
        mapping_dict = {k.lower(): v for k, v in mapping_dict.items()}

    # Apply standardization
    df[column_name] = df[column_name].map(mapping_dict).fillna(df[column_name])

    # Post-standardization report
    standardized_values = df[column_name].value_counts(dropna=False)
    print(f"Remaining unique values after standardization: {len(standardized_values)}")

    # Show changes
    changes = set(original_values.index) - set(standardized_values.index)
    if changes:
        print(f"\nValues that were standardized: {len(changes)}")
        # Convert all values to strings for sorting
        sorted_changes = sorted([str(x) for x in changes])[:10]
        for val in sorted_changes:
            lookup_val = val.lower() if not case_sensitive else val
            new_val = mapping_dict.get(lookup_val, val)
            print(f" - {val} → {new_val}")

    return df

# Example: Country standardization
country_mapping = {
    # Common abbreviations
    'usa': 'United States',
    'us': 'United States',
    'united states of america': 'United States',
    'u.s.': 'United States',
    'uk': 'United Kingdom',
    'u.k.': 'United Kingdom',
    'great britain': 'United Kingdom',
    # Common misspellings
    'united kindom': 'United Kingdom',
    'america': 'United States',
    # Handle numeric values that might appear as strings
    'nan': None  # Handle NaN values converted to string
}

# Example: Education Level standardization
edlevel_mapping = {
    'bachelor\'s degree': 'Bachelor\'s degree',
    'bachelors': 'Bachelor\'s degree',
    'bs': 'Bachelor\'s degree',
    'master\'s degree': 'Master\'s degree',
    'masters': 'Master\'s degree',
    'ms': 'Master\'s degree',
    'phd': 'Doctoral degree',
    'doctoral': 'Doctoral degree',
    'some college': 'Some college/university study',
    'college': 'Some college/university study',
    'nan': None  # Handle NaN values converted to string
}

# Apply standardization
df = standardize_column(df, 'Country', country_mapping, case_sensitive=False)
df = standardize_column(df, 'EdLevel', edlevel_mapping, case_sensitive=False)

# Show results
print("\nSample of standardized data:")
print(df[['Country', 'EdLevel']].head(10))
```

```python
# 5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.¶
```

```python
import pandas as pd
import numpy as np

# Load the dataset
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# 5.2 Impute missing values in numerical columns
def impute_numerical(df, columns=None, strategy='median', report=True):
    """
    Impute missing values in numerical columns using specified strategy

    Parameters:
    - df: DataFrame
    - columns: List of columns to impute (None for all numerical columns)
    - strategy: 'mean', 'median', or 'zero'
    - report: Whether to show before/after stats
    """
    if columns is None:
        # Select all numerical columns
        numerical_cols = df.select_dtypes(include=[np.number]).columns
    else:
        numerical_cols = columns

    if report:
        print("=== Before Imputation ===")
        print("Columns with missing values:")
        print(df[numerical_cols].isna().sum()[df[numerical_cols].isna().sum() > 0])

    for col in numerical_cols:
        if df[col].isna().any():
            if strategy == 'mean':
                impute_value = df[col].mean()
            elif strategy == 'median':
                impute_value = df[col].median()
            elif strategy == 'zero':
                impute_value = 0
            else:
                raise ValueError("Strategy must be 'mean', 'median', or 'zero'")

            # Store imputation value and count
            impute_count = df[col].isna().sum()
            df[col].fillna(impute_value, inplace=True)

            if report:
                print(f"\nImputed {impute_count} missing values in '{col}' with {strategy} value: {impute_value:.2f}")

    if report:
        print("\n=== After Imputation ===")
        print("Remaining missing values in numerical columns:")
        print(df[numerical_cols].isna().sum()[df[numerical_cols].isna().sum() > 0])

    return df

# Example usage - impute specific column with median
df = impute_numerical(df, columns=['ConvertedCompYearly'], strategy='median')

# Example usage - impute all numerical columns with mean
# df = impute_numerical(df, strategy='mean')

# Show results
print("\nSample data after imputation:")
print(df[['ConvertedCompYearly']].head())
```

```python
# <h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>
```

```python
import pandas as pd

# Load the dataset
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# 5.3 Impute missing values in categorical columns with mode
def impute_categorical(df, columns=None, report=True):
    """
    Impute missing values in categorical columns using mode (most frequent value)

    Parameters:
    - df: DataFrame
    - columns: List of columns to impute (None for all categorical columns)
    - report: Whether to show before/after stats
    """
    if columns is None:
        # Select all categorical columns (object and category dtypes)
        categorical_cols = df.select_dtypes(include=['object', 'category']).columns
    else:
        categorical_cols = columns

    if report:
        print("=== Before Imputation ===")
        print("Categorical columns with missing values:")
        print(df[categorical_cols].isna().sum()[df[categorical_cols].isna().sum() > 0])

    for col in categorical_cols:
        if df[col].isna().any():
            # Get most frequent value (mode)
            mode_value = df[col].mode()[0]
            impute_count = df[col].isna().sum()

            # Impute missing values
            df[col].fillna(mode_value, inplace=True)

            if report:
                print(f"\nImputed {impute_count} missing values in '{col}' with mode: '{mode_value}'")

    if report:
        print("\n=== After Imputation ===")
        print("Remaining missing values in categorical columns:")
        print(df[categorical_cols].isna().sum()[df[categorical_cols].isna().sum() > 0])

    return df

# Example usage - impute specific column
df = impute_categorical(df, columns=['RemoteWork'])

# Example usage - impute all categorical columns
# df = impute_categorical(df)

# Show results
print("\nSample data after imputation:")
print(df[['RemoteWork']].head())

# Verify the imputation
print("\nValue counts after imputation:")
print(df['RemoteWork'].value_counts(dropna=False))
```

```python
# 6. Feature Scaling and Transformation
# 6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# 6.1 Min-Max Scaling for ConvertedCompYearly
def min_max_scale_column(df, column_name):
    """
    Apply Min-Max scaling to a numerical column
    """
    print(f"\n=== Min-Max Scaling for {column_name} ===")

    # Check original stats
    original_min = df[column_name].min()
    original_max = df[column_name].max()
    print(f"Original range: [{original_min}, {original_max}]")

    # Handle missing values (drop for scaling)
    non_missing = df[column_name].dropna()

    # Initialize and fit scaler
    scaler = MinMaxScaler()
    scaled_values = scaler.fit_transform(non_missing.values.reshape(-1, 1))

    # Create new column for scaled values
    new_col_name = f"{column_name}_MinMaxScaled"
    df[new_col_name] = pd.NA  # Initialize with missing values
    df.loc[non_missing.index, new_col_name] = scaled_values.flatten()

    # Verify results
    print(f"New range: [{df[new_col_name].min():.2f}, {df[new_col_name].max():.2f}]")
    print(f"Missing values preserved: {df[column_name].isna().sum()}")

    return df, scaler

# Apply scaling
df, scaler = min_max_scale_column(df, 'ConvertedCompYearly')

# Show results
print("\nSample of original vs scaled values:")
sample = df[['ConvertedCompYearly', 'ConvertedCompYearly_MinMaxScaled']].dropna().head()

# Corrected float formatting
pd.set_option('display.float_format', lambda x: '%.3f' % x)
print(sample.to_string())

# Optional: Visual comparison
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
df['ConvertedCompYearly'].hist(bins=30)
plt.title('Original Values')

plt.subplot(1, 2, 2)
df['ConvertedCompYearly_MinMaxScaled'].hist(bins=30)
plt.title('Min-Max Scaled Values')

plt.tight_layout()
plt.show()
```

```python
# <h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Load the dataset
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# 6.2 Log-transform ConvertedCompYearly
def log_transform_column(df, column_name, add_constant=1):
    """
    Apply log transformation to reduce skewness
    add_constant: Small value to add to handle zeros (default 1)
    """
    print(f"\n=== Log Transformation for {column_name} ===")

    # Handle missing values
    non_missing = df[column_name].dropna()

    # Check original skewness
    original_skew = stats.skew(non_missing)
    print(f"Original skewness: {original_skew:.2f}")

    # Apply log transformation (add small constant to handle zeros)
    new_col_name = f"{column_name}_LogTransformed"
    df[new_col_name] = np.log(non_missing + add_constant)

    # Check new skewness
    transformed_skew = stats.skew(df[new_col_name].dropna())
    print(f"Transformed skewness: {transformed_skew:.2f}")
    print(f"Skewness reduction: {abs(original_skew - transformed_skew):.2f}")

    # Visual comparison
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.hist(non_missing, bins=50)
    plt.title(f'Original {column_name}\nSkewness: {original_skew:.2f}')

    plt.subplot(1, 2, 2)
    plt.hist(df[new_col_name].dropna(), bins=50)
    plt.title(f'Log-Transformed {column_name}\nSkewness: {transformed_skew:.2f}')

    plt.tight_layout()
    plt.show()

    return df

# Apply log transformation (adding 1 to handle zeros)
df = log_transform_column(df, 'ConvertedCompYearly', add_constant=1)

# Show sample results
print("\nSample of transformed values:")
sample = df[['ConvertedCompYearly', 'ConvertedCompYearly_LogTransformed']].dropna().head()
pd.set_option('display.float_format', lambda x: '%.3f' % x)
print(sample.to_string())
```

```python
# 7. Feature Engineering
# 7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:
```

```python
import pandas as pd
import numpy as np

# Load the dataset
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# 7.1 Create ExperienceLevel from YearsCodePro
def create_experience_level(df):
    """
    Categorize professional experience based on YearsCodePro
    """
    print("\n=== Creating ExperienceLevel column ===")

    # Convert YearsCodePro to numeric, coercing errors to NaN
    df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

    # Check cleaned distribution
    print("YearsCodePro distribution after cleaning:")
    print(df['YearsCodePro'].describe())

    # Define experience level bins and labels
    bins = [0, 2, 5, 10, 20, np.inf]
    labels = ['Entry-level (0-2)', 'Junior (3-5)', 'Mid-level (6-10)',
              'Senior (11-20)', 'Principal (20+)']

    # Create new column
    df['ExperienceLevel'] = pd.cut(
        df['YearsCodePro'],
        bins=bins,
        labels=labels,
        right=False,
        include_lowest=True
    )

    # Handle missing values
    df['ExperienceLevel'] = df['ExperienceLevel'].cat.add_categories('Missing')
    df['ExperienceLevel'].fillna('Missing', inplace=True)

    # Show results
    print("\nExperienceLevel distribution:")
    print(df['ExperienceLevel'].value_counts().sort_index())

    return df

# Apply the function
df = create_experience_level(df)

# Show sample results
print("\nSample of engineered feature:")
sample_cols = ['YearsCodePro', 'ExperienceLevel']
print(df[sample_cols].dropna(subset=['YearsCodePro']).head(10).to_string())
```
